```python
import os
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import csv
```

```python
D = 27.0 # m
R = D/2.0
rhoFac_neut = 1.02/1.2256
rhoFac_unst = 1.09/1.2256
perf_file_FAST = os.path.join("v1","Simulated_data", "RotorPerformances_OF2" +".csv")
perf_FAST      = pd.read_csv(perf_file_FAST, sep='\s+', skiprows=1,
                     names=['wind', 'rpm', 'pitch', 'TSR', 'cp', 'ct', 'power', 'FlapBM'])
print(perf_FAST)
print('FAST FlapBM interpolated at 6.7 and 8.7 m/s:')
print(np.interp([6.7, 8.7], perf_FAST['wind'],perf_FAST['FlapBM']))
```

```python
perf_file_GW_neutral_old = os.path.join("V27","Simulated_data", "RotorPerformances_GW_neutral2" +".csv")
perf_GW_neutral_old      = pd.read_csv(perf_file_GW_neutral_old, sep='\s+', skiprows=1,
                     names=['uRef', 'time', 'power', 'thrust', 'wind', 'windDir', 'torque', 'rpm', 'pitch', 'yawOffset', 'NCYaw'])
#print(perf_GW_neutral_old)
```

```python
perf_file_GW_neutral = os.path.join("V27","Simulated_data", "RotorPerformances_GW_neutral5b" +".csv")
perf_GW_neutral      = pd.read_csv(perf_file_GW_neutral, sep='\s+', skiprows=1,
                     names=['uRef', 'time', 'power', 'thrust', 'wind', 'Uhub','windDir','windDirYaw','windDirHub','windDirYawHub','torque', 'rpm', 'pitch', 'yawOffset', 'NCYaw'])
print(perf_GW_neutral)
```

```python
perf_file_GW_unstable_old = os.path.join("V27","Simulated_data", "RotorPerformances_GW_unstable2" +".csv")
perf_GW_unstable_old      = pd.read_csv(perf_file_GW_unstable_old, sep='\s+', skiprows=1,
                     names=['uRef', 'time', 'power', 'thrust', 'wind', 'windDir', 'torque', 'rpm', 'pitch', 'yawOffset', 'NCYaw'])
#print(perf_GW_unstable_old)
```

```python
perf_file_GW_unstable = os.path.join("V27","Simulated_data", "RotorPerformances_GW_unstable5b" +".csv")
perf_GW_unstable      = pd.read_csv(perf_file_GW_unstable, sep='\s+', skiprows=1,
                     names=['uRef', 'time', 'power', 'thrust', 'wind', 'Uhub','windDir','windDirYaw','windDirHub','windDirYawHub','torque', 'rpm', 'pitch', 'yawOffset', 'NCYaw'])
print(perf_GW_unstable)
```

```python
perf_file_GW_neutral_mom = os.path.join("V27","Simulated_data", "RotorPerformances_GW_neutral5b_mom" +".csv")
perf_GW_neutral_mom      = pd.read_csv(perf_file_GW_neutral_mom, sep='\s+', skiprows=1,
                     names=['uRef', 'flapBM'])
print(perf_GW_neutral_mom)
print(perf_GW_neutral_mom['flapBM']*rhoFac_neut)
```

```python
perf_file_GW_unstable_mom = os.path.join("V27","Simulated_data", "RotorPerformances_GW_unstable5b_mom" +".csv")
perf_GW_unstable_mom      = pd.read_csv(perf_file_GW_unstable_mom, sep='\s+', skiprows=1,
                           names=['uRef', 'flapBM'])
print(perf_GW_unstable_mom)
print(perf_GW_unstable_mom['flapBM']*rhoFac_unst)
```

```python
perf_file_GW_thrust = os.path.join("V27","Simulated_data", "FlapBM_HandComputation" +".csv")
perf_GW_thrust      = pd.read_csv(perf_file_GW_thrust, sep='\s+', skiprows=0,
                                 names=['rLoc', 'th_neut', 'th_unst'])
#print(perf_GW_thrust)
flapBM_neut = (perf_GW_thrust['rLoc']*perf_GW_thrust['th_neut']).sum()*rhoFac_neut*34/1000.0
flapBM_unst = (perf_GW_thrust['rLoc']*perf_GW_thrust['th_unst']).sum()*rhoFac_unst*20/1000.0
print (flapBM_neut)
print (flapBM_unst)
```

```python
# Plot RPM
plt.figure(1) 
plt.plot(perf_FAST['wind'],perf_FAST['rpm'], '-b', linewidth=2, label= 'FAST')
plt.plot(perf_GW_neutral_old['wind'],perf_GW_neutral_old['rpm'], '-og', linewidth=2,label= 'neutral, unref')
plt.plot(perf_GW_unstable_old['wind'],perf_GW_unstable_old['rpm'], '-sr', linewidth=2,label= 'unstable, unref')
plt.plot(perf_GW_neutral['wind'],perf_GW_neutral['rpm'], '-^k', linewidth=2,label= 'neutral, ref')
plt.plot(perf_GW_unstable['wind'],perf_GW_unstable['rpm'], '-dm', linewidth=2,label= 'unstable, ref')
plt.grid()
plt.legend(loc='best')
plt.xlabel('Wind speed (m/s)')
plt.ylabel('RPM')
plt.savefig('RPM_vs_windspeed.png')
plt.show()
```

```python
# Plot TSR
plt.figure(2) 
plt.plot(perf_FAST['wind'],perf_FAST['TSR'], '-b', linewidth=2, label= 'FAST')
plt.plot(perf_GW_neutral_old['wind'],(perf_GW_neutral_old['rpm']*R/perf_GW_neutral_old['wind'])*np.pi/30.0, '-og', linewidth=2,label= 'neutral, unref')
plt.plot(perf_GW_unstable_old['wind'],(perf_GW_unstable_old['rpm']*R/perf_GW_unstable_old['wind'])*np.pi/30.0, '-sr', linewidth=2,label= 'unstable, unref')
plt.plot(perf_GW_neutral['wind'],(perf_GW_neutral['rpm']*R/perf_GW_neutral['wind'])*np.pi/30.0, '-^k', linewidth=2,label= 'neutral, ref')
plt.plot(perf_GW_unstable['wind'],(perf_GW_unstable['rpm']*R/perf_GW_unstable['wind'])*np.pi/30.0, '-dm', linewidth=2,label= 'unstable, ref')
plt.grid()
plt.grid()
plt.legend(loc='best')
plt.xlabel('Wind speed (m/s)')
plt.ylabel('TSR')
plt.savefig('TSR_vs_windspeed.png')
plt.show()
```

```python
# Plot Power
plt.figure(3) 
plt.plot(perf_FAST['wind'],perf_FAST['power'], '-b', linewidth=2, label= 'FAST')
plt.plot(perf_GW_neutral_old['wind'],perf_GW_neutral_old['power'], '-og', linewidth=2,label= 'neutral, unref')
plt.plot(perf_GW_unstable_old['wind'],perf_GW_unstable_old['power'], '-sr', linewidth=2,label= 'unstable, unref')
plt.plot(perf_GW_neutral['wind'],perf_GW_neutral['power'], '-^k', linewidth=2,label= 'neutral, ref')
plt.plot(perf_GW_unstable['wind'],perf_GW_unstable['power'], '-dm', linewidth=2,label= 'unstable, ref')
plt.plot(perf_GW_neutral['wind'],perf_GW_neutral['power']*rhoFac_neut, '-.k', linewidth=2,label= 'neutral, ref, density-scaled')
plt.plot(perf_GW_unstable['wind'],perf_GW_unstable['power']*rhoFac_unst, '-.m', linewidth=2,label= 'unstable, ref, density-scaled')
plt.grid()
plt.legend(loc='best')
plt.xlabel('Wind speed (m/s)')
plt.ylabel('Power (kW)')
plt.savefig('Power_vs_windspeed.png')
plt.show()
```

```python
# Plot Torque
plt.figure(3) 
plt.plot(perf_FAST['wind'],perf_FAST['power']/(perf_FAST['rpm']*np.pi/30.0), '-b', linewidth=2, label= 'FAST')
plt.plot(perf_GW_neutral['wind'],perf_GW_neutral['torque']*rhoFac_neut, '-.k', linewidth=2,label= 'neutral, ref, density-scaled')
plt.plot(perf_GW_unstable['wind'],perf_GW_unstable['torque']*rhoFac_unst, '-.m', linewidth=2,label= 'unstable, ref, density-scaled')
plt.grid()
plt.legend(loc='best')
plt.xlabel('Wind speed (m/s)')
plt.ylabel('Torque (kN-m)')
plt.savefig('Torque_vs_windspeed.png')
plt.show()
```

```python
# Plot Cp
Cp_FAST_computed = 1000.0*perf_FAST['power']/(0.5*1.1*np.pi*R*R*perf_FAST['wind']*perf_FAST['wind']*perf_FAST['wind'])
#print (Cp_FAST_computed)
Cp_GW_computed_neutral = 1000.0*perf_GW_neutral['power']/(0.5*1.2256*np.pi*R*R*perf_GW_neutral['wind']*perf_GW_neutral['wind']*perf_GW_neutral['wind'])
#print (Cp_GW_computed_neutral)
Cp_GW_computed_unstable = 1000.0*perf_GW_unstable['power']/(0.5*1.2256*np.pi*R*R*perf_GW_unstable['wind']*perf_GW_unstable['wind']*perf_GW_unstable['wind'])
#print (Cp_GW_computed_unstable)
plt.figure(4) 
plt.plot(perf_FAST['wind'],perf_FAST['cp'], '-b', linewidth=2, label= 'FAST')
#plt.plot(perf_FAST['wind'],Cp_FAST_computed, '-k', linewidth=2, label= 'FAST: derived')
plt.plot(perf_GW_neutral['wind'],Cp_GW_computed_neutral, '-og', linewidth=2,label= 'neutral, ref')
plt.plot(perf_GW_unstable['wind'],Cp_GW_computed_unstable, '-sr', linewidth=2,label= 'unstable, ref')
plt.grid()
plt.legend(loc='lower left')
plt.xlabel('Wind speed (m/s)')
plt.ylabel('Power coefficient, Cp')
plt.savefig('Cp_vs_windspeed.png')
plt.show()
```

```python
# Plot Ct
Ct_GW_computed_neutral = 1000.0*perf_GW_neutral['thrust']/(0.5*1.2256*np.pi*R*R*perf_GW_neutral['wind']*perf_GW_neutral['wind'])
#print (Ct_GW_computed_neutral)
Ct_GW_computed_unstable = 1000.0*perf_GW_unstable['thrust']/(0.5*1.2256*np.pi*R*R*perf_GW_unstable['wind']*perf_GW_unstable['wind'])
#print (Ct_GW_computed_unstable)
plt.figure(4) 
plt.plot(perf_FAST['wind'],perf_FAST['ct'], '-b', linewidth=2, label= 'FAST')
#plt.plot(perf_FAST['wind'],Cp_FAST_computed, '-k', linewidth=2, label= 'FAST: derived')
plt.plot(perf_GW_neutral['wind'],Ct_GW_computed_neutral, '-og', linewidth=2,label= 'neutral, ref')
plt.plot(perf_GW_unstable['wind'],Ct_GW_computed_unstable, '-sr', linewidth=2,label= 'unstable, ref')
plt.grid()
plt.legend(loc='lower left')
plt.xlabel('Wind speed (m/s)')
plt.ylabel('Thrust coefficient, Ct')
plt.savefig('Ct_vs_windspeed.png')
plt.show()
```
